The `mlxtend` library provides an ExhaustiveFeatureSelector class for this purpose. You can specify the min and max number of features.

```python

from mlxtend import ExhaustiveFeatureSelector as EFS

model = LogisticRegression(max_iter=500)

efs = EFS(
    model,
    min_features=1,
    max_features=3,  # limit max features to reduce computation
    scoring='accuracy',
    cv=5,
    print_progress=True
)

efs.fit(X_train, y_train)

print("Best features:", efs.best_feature_names_)
print("Best CV score:", efs.best_score_)


```
